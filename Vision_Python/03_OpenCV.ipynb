```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Con esto se lee la imagen mediante la libreria opencv 
image = cv2.imread('./images/perro.jpg')
```

```python
type(image)
```

```python
# Se muetra las dimensiones de la imagen convertida en arreglo
image.shape
```

```python
plt.imshow(image) 
# No se ve del mismo color al original por la integracion matplotlib y cv2
# OpenCV --> BGR
# Matplotlib --> RGB
```

```python
# Se realiza la conversion de BGR a RGB. Converts an image from one color space to another.
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
# Esta es una forma mas rapida de hacer uso de la imagen
image = cv2.imread('./images/perro.jpg', cv2.IMREAD_COLOR_RGB)
plt.imshow(image)
```

```python
# La imagen se convierte a una imagen de un solo canal en escala de grises
image_wandb = cv2.imread('./images/perro.jpg', cv2.IMREAD_GRAYSCALE)
# Se valida que la imagen se encuentra solo en una dimension (canal)
print(image_wandb.shape)
```

```python
# Se visualiza la imagen aplicando una capa de colores en escala de grises
plt.imshow(image_wandb, cmap='gray')
```
