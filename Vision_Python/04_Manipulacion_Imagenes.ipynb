```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imagen = cv2.imread('./images/perro.jpg', cv2.IMREAD_COLOR_RGB)
```

```python
plt.imshow(imagen)
```

```python
imagen.shape
```

```python
imagen_2 = cv2.resize(imagen, (306, 279))
```

```python
plt.imshow(imagen_2)
```

```python
ratio_ancho = 0.5
ratio_alto = 0.5
imagen_3 = cv2.resize(imagen, None, None, ratio_alto, ratio_ancho)
```

```python
plt.imshow(imagen_3)
```

```python
imagen_3.shape
```

```python
imagen_2 = cv2.flip(imagen_2, 1)
plt.imshow(imagen_2)
```

```python
# Los valores permitidos como segundo parametro del flip son
# 0 > para girar en ambos axis
# 0 para girar en axis x
# 0 < para girar en axis y
imagen_2 = cv2.flip(imagen_2, 0)
plt.imshow(imagen_2)
```

```python
imagen_2 = cv2.cvtColor(imagen_2, cv2.COLOR_RGB2BGR)
cv2.imwrite('./images/perro_modificado.jpg',imagen_2)
```
