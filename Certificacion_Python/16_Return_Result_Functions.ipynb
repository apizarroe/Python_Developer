```python
# Usando el return sin expresion
# Al enviar el return sin parametros en el momento de la ejecuci칩n
# causa el termino de la instrucci칩n desde ese punto. 
def happy_new_year(wishes = True):
    print("Three...")
    print("Two...")
    print("One...")
    if not wishes:
        return
    print("Happy New Year!")
```

```python
print("Ejecucion sin return")
happy_new_year()
print("Ejecucion con return")
happy_new_year(wishes=False)
```

```python
# Usando el return con una expresion
# Existen dos consecuencias de usar esto
# it causes the immediate termination of the function's execution
# moreover, the function will evaluate the expression's value and will return it, as the function's result.

def boring_function():
    return 123
 
x = boring_function()
 
print("The boring_function has returned its result. It's:", x)
```

```python
# Efectos y resultados: Listas y funciones
# Una listas puede ser enviada a una funcion como un argumento
# Como en este ejemplo:
def list_sum(lst):
    s = 0
 
    for elem in lst:
        s += elem
 
    return s

print(list_sum([5, 4, 3]))

# Aca se tiene un error en la invocacion pues espera una lista
# print(list_sum(5))
print(list_sum([5])) # Esta es la forma correcta de enviar una lista de un elemento
```

```python
# Una lista tambien puede ser el resultado de una funcion
def strange_list_fun(n):
    strange_list = []
    
    for i in range(n):
        strange_list.insert(0, i)
    
    return strange_list

print(strange_list_fun(5))
```

```python
# LAB3: Escribiendo funciones

def is_year_leap(year):
    if (year % 4 == 0):
        if (year % 400 == 0):
            return True
        elif (year % 100 == 0):
            return False
        else:
            return True
    else:
        return False

def days_in_month(year, month):
    year_leap = is_year_leap(year)
    days = 0
    if (month in [1, 3, 5, 7, 8, 10, 12]):
        days = 31
    elif (month in [4, 6, 9, 11]):
        days = 30
    elif (month in [2]):
        if year_leap == True:
            days = 29
        else:
            days = 28
    return days

def day_of_year(year, month, day):
    days_total = 0
    for i in range(1, month):
        days_total += days_in_month(year, i)
    if day <= days_in_month(year, month):
        days_total += day
    else:
        days_total = None

    return days_total

print(day_of_year(1900, 2, 30))
```

```python
# LAB 4: Numeros Primos
def is_prime(num):
    if num < 2:
        return False
    if num in (2, 3):
        return True
    if num % 2 == 0:
        return False
    raiz = int(num ** 0.5) + 1
    for i in range(3, raiz, 2):
        if num % i == 0:
            return False
    return True

for i in range(1, 20):
    if is_prime(i + 1):
        print(i + 1, end=" ")
print()
```
