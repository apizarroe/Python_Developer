```python
# Listas en Listas
# Las listas ademas de contener valores literales tambien puede contener otras listas
# Haciendo un simil de un tablero de ajedrez donde se tienen un arreglo de 8 x 8
# donde el tablero es representado por filas y cada una de ellas es una lista
WHITE_PAWN = "white_pawn"
row = []
for i in range(8):
    row.append(WHITE_PAWN)
print(row)
```

```python
# Comprension de Listas
# El mismo efecto se logra con la comprension de listas, 
# una sintaxis especial en Python para llenar listas de forma masiva
# Esto es una lista creada durante la ejecución del programa y no descrita de forma estática
# Sigue la siguiente sintaxys [expression for element in list if conditional]
WHITE_PAWN = "white_pawn"
row = [WHITE_PAWN for i in range(8)]
print(row)
```

```python
# Esta conformada por la data que será ingestada en durante la ejecución
# y la clausula que define cuantas veces ocurrirá dentro de la lista
squares = [x ** 2 for x in range(10)]
print(squares)
```

```python
# Ejemplo con las potencias de 2
twos = [2 ** i for i in range(8)]
print(twos)
```

```python
# En este caso no se realiza operacion alguna para calculo
# Sin embargo se utiliza la sentencia if para el filtrado de datos
odds = [x for x in squares if x % 2 != 0 ]
print(odds)
```

```python
# Se crea una lista de lista representando el tablero completo de ajedrez
# Primero se crea una lista con 8 espacios vacios en la lista row, 
# luego la lista se repite la misma cantidad de veces para la lista de listas board
board = []
EMPTY = "empty"

for i in range(8):
    row = [EMPTY for i in range(8)]
    board.append(row)

display(board)
# La variable board es un arreglo de dos dimensiones.
# ESto tambien puede ser llamado una matriz
```

```python
# La comprension de listas puede ser anidad, de la siguiente manera:
board = [[EMPTY for i in range(8)] for j in range(8)]
display(board)
# La parte interna crea una fila y la parte externa crea una lista de filas
print(board[2][3])
```

```python
# Se colocan las torres en sus posiciones
board[0][0] = 'ROOK'
board[0][7] = 'ROOK'
board[7][0] = 'ROOK'
board[7][7] = 'ROOK'

# Se coloca el caballo
board[4][2] = 'KNIGHT'

# Se coloca un peon
board[3][4] = 'PAWN'

board
```

```python
# Naturaleza Multidimensional de las listas: Aplicaciones Avanzadas
# Para usar elementos en listas de dos dimensiones, se debe usar dos cordenadas:
# Una vertical (numero de fila)
# Una horizontal (numero de columna)
# Imaginamos que requerimos almacenar los valores de temperatura de un sensor,
# suministrados cada hora durante un mes, con lo cual se debe almacenar un total
# de 24 * 32 = 744 valores, los cuales son float pues los valores requieren una precision de 0.1 °C.
# Cada valor en el la fila sera un valor de temperatura en una hora 
# y cada fila representa un dia en el mes
```

```python
temps = [[0.0 for h in range(24)] for d in range(31)]
temps
```

```python
# Vamos a determinar el promedio de la temperatura en la tarde.
# Sumamos las 31 lecturas al medio dia y dividimos la suma entre 31.
# Se puede asumir que la temperatura de la media noche esta en la primera posicion.

temps = [[0.0 for h in range(24)] for d in range(31)]
#
# The matrix is magically updated here.
#
 
total = 0.0
 
for day in temps:
    total += day[11]
 
average = total / 31
 
print("Average temperature at noon:", average)

# La variable dia usada en el loop no es escalar, cada paso de la matrix temps
# asigna el valor de la fila a la variable day la cual es una lista.
# Luego toma el valor indexado con 11 que es el valor del medio dia,
```

```python
# En este ejemplo la variable day itera a travez de todas las filas en la matriz temps
# y la variable temp itera para traer la mediciones de un dia

temps = [[0.0 for h in range(24)] for d in range(31)]
#
# The matrix is magically updated here.
#
temps = [[100,50,75],[86,128,136],[30,45,78]]
 
highest = -100.0
 
for day in temps:
    for temp in day:
        if temp > highest:
            highest = temp
 
print("The highest temperature was:", highest)
```

```python
# Se contabiliza todos los dias que tuvieron >20°C al medio dia
temps = [[0.0 for h in range(24)] for d in range(31)]
#
# The matrix is magically updated here.
#
temps = [[10,13,21],[10,13,19],[10,13,23]]

hot_days = 0
 
for day in temps:
    if day[2] > 20.0:
        hot_days += 1
 
print(hot_days, "days were hot.")
```

```python
# Python no tiene limitacion de profundida de inclusion de listas en listas,
# Aca un ejemplo de lista de tres dimensiones
rooms = [[[False for r in range(20)] for f in range(15)] for t in range(3)]
display(rooms)
# Este seria un ejemplo de una representacion de un hotel de 3 edificios,
# 15 pisos en cada uno y 20 habitaciones en cada piso.
# Para esto se necesita recolectar la información de las habitaciones libres (True/False)
```

```python
# El primer indice, representa al edificio (0 - 2)
# El segundo indice, representa al piso (0 - 14)
# El tercer indice, representa a la habitacion (0 - 19)

# Se debe reservar una habitación
rooms[1][9][13] = True
# Se debe liberar una habitación
rooms[0][4][1] = False
display(rooms)
```

```python
vacancy = 0
# Por defecto todas la habitaciones se encuentran disponibles
# Habitaciones Ocupadas
rooms[2][14][2] = True
rooms[2][14][3] = True
rooms[2][14][10] = True
rooms[2][14][15] = True

# La variable vacancy, muestra la cantidad de habitaciones disponibles
for room_number in range(20):
    if not rooms[2][14][room_number]:
        vacancy += 1

print(vacancy)
```
